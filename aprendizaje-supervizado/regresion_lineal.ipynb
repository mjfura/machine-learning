```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(0)
sample_size = 100
x = np.sort(5*np.random.rand(sample_size, 1), axis=0)
y = 2+3*x + np.random.randn(sample_size, 1)
```

```python
reg = LinearRegression().fit(x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color='black')
plt.show()
```

```python
print(f'Coeficientes: {reg.coef_[0]}')
print(f'Intercepto: {reg.intercept_[0]}')
```

Usaremos ahora 2 variables predictoras

```python
x_1 = np.sort(5*np.random.rand(sample_size, 1), axis=0)
x_2 = np.sort(5*np.random.rand(sample_size, 1), axis=0)
x_2d = np.hstack((x_1, x_2))
```

```python
y_2 = 2 + 3*x_1 + 4*x_2 + np.random.randn(sample_size, 1)
```

```python
reg_2 = LinearRegression().fit(x_2d, y_2)
```

```python
model = LinearRegression()
model.fit(x_2d, y_2)

# Crear una malla de puntos para graficar el plano de prediccin
x1_range = np.linspace(x_2d[:, 0].min(), x_2d[:, 0].max(), 10)
x2_range = np.linspace(x_2d[:, 1].min(), x_2d[:, 1].max(), 10)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)
y_pred = model.predict(np.c_[x1_grid.ravel(), x2_grid.ravel()]).reshape(x1_grid.shape)

# Crear una grfica 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Graficar los puntos de datos
ax.scatter(x_2d[:, 0], x_2d[:, 1], y_2, c="r", marker="o")

# Graficar el plano de prediccin
ax.plot_surface(x1_grid, x2_grid, y_pred, color="b", alpha=0.5)

# Etiquetas de los ejes
ax.set_xlabel("Variable Predictora 1")
ax.set_ylabel("Variable Predictora 2")
ax.set_zlabel("Variable de Respuesta")

plt.show()
```

```python
print(f'Coeficientes: {model.coef_}')
print(f'Intercepto: {model.intercept_}')
```
