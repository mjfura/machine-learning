```python
from scipy.stats import pearsonr
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
sample_size=250
x=np.random.uniform(-5,5,sample_size)
y1=x+np.random.normal(0,1,sample_size)
y2=x**2+np.random.normal(0,1,sample_size)
y3=5*np.sin(0.5*np.pi*x) + np.random.normal(0,1,sample_size)
y4=-1*x+np.random.normal(0,1,sample_size)
```

```python
p_1=pearsonr(x,y1)[0]
p_2=pearsonr(x,y2)[0]
p_3=pearsonr(x,y3)[0]
p_4=pearsonr(x,y4)[0]
```

```python
fig,ax=plt.subplots(2,2,figsize=(15,10))
fig.subplots_adjust(hspace=0.3)
ax[0,0].scatter(x,y1)
ax[0,0].set_title('Score: {:.2f}'.format(p_1))
ax[0,1].scatter(x,y2)
ax[0,1].set_title('Score: {:.2f}'.format(p_2))
ax[1,0].scatter(x,y3)
ax[1,0].set_title('Score: {:.2f}'.format(p_3))
ax[1,1].scatter(x,y4)
ax[1,1].set_title('Score: {:.2f}'.format(p_4))
plt.show()
```

```python
np.random.seed(0)
sample_size=80
x=np.sort(5*np.random.rand(sample_size,1),axis=0)
y=np.sin(x) + 0.1*np.random.randn(sample_size,1)
```

```python
reg_linear = LinearRegression().fit(x,y)
```

```python
x_poly = np.hstack((x,x**2,x**3,x**4))
reg_poly = LinearRegression().fit(x_poly,y)
x_grid = np.linspace(0,5,100).reshape(-1,1)
x_grid_poly = np.hstack((x_grid,x_grid**2,x_grid**3,x_grid**4))
plt.scatter(x,y,color='black')
plt.plot(x_grid,reg_linear.predict(x_grid),label='Linear', color='red')
plt.plot(x_grid,reg_poly.predict(x_grid_poly),label='Polynomial', color='blue')
plt.legend()
plt.show()
```
